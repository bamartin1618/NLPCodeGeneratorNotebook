```python
import os
os.environ['CUDA_LAUNCH_BLOCKING'] = "1"

import logging

logging.basicConfig(
        format="%(asctime)s — %(levelname)s — %(name)s — %(message)s",
        datefmt="%m/%d/%Y %H:%M:%S",
        level=logging.INFO
)

from aitextgen import aitextgen
from aitextgen.TokenDataset import TokenDataset
from aitextgen.utils import build_gpt2_config
from aitextgen.tokenizers import train_tokenizer
import numpy as np
```

```python
import os

print(os.listdir(os.curdir))
```

```python
file_name = 'joke_dataset.txt'

train_tokenizer(file_name)
tokenizer_file = "aitextgen.tokenizer.json"

training_dataset = TokenDataset(file_name, tokenizer_file=tokenizer_file, save_cache=True)
training_file = 'dataset_cache.tar.gz'
```

```python
ai_tuned = aitextgen(model="EleutherAI/gpt-neo-125M", to_gpu=True)
```

```python
ai_tuned.train(training_file,
         line_by_line = False,
         from_cache = True,
         num_steps = 1000,
         generate_every = 500,
         save_every = 500,
         save_gdrive = False,
         learning_rate = 5e-3,
         batch_size = 1,
         output_dir = "tuned_model")
```
